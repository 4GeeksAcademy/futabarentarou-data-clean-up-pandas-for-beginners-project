#### Ejercicio 01. Análisis de datos básico con estructuras nativas de python.
Crea una lista con las calificaciones de 5 estudiantes: [8.5, 9.2, 7.8, 8.9, 10].

- Calcula el promedio de las calificaciones.

```python
calificaciones = [8.5, 9.2, 7.8, 8.9, 10]
```

```python
promedio = sum(calificaciones) / len(calificaciones)

print("El promedio es:", promedio)
```

- Encuentra la calificación más alta y la más baja.

```python
alta = max(calificaciones)
baja = min(calificaciones)

print("La calificación más alta es:", alta)
print("La calificación más baja es:", baja)
```

#### Ejercicio 01. ¿Cuál es la casa más cara en todo el dataset? (★☆☆)

Imprime la dirección y el precio de la casa seleccionada. Por ejemplo:

`La casa con dirección en Calle del Prado, Nº20 es la más cara y su precio es de 5000000 USD`

```python
casamascara = ds.loc[ds["price"].idxmax()]

print(f'La casa mas cara es {casamascara["address"]}, su precio es de {casamascara["price"]} USD')
```

#### Ejercicio 02. ¿Cuál es la casa más barata del dataset? (★☆☆)

Este ejercicio es similar al ejercicio anterior, solo que ahora buscamos la casa con el precio más bajo. Recuerda imprimir la dirección y el precio de la casa seleccionada. Por ejemplo:

`La casa con dirección en Calle Alcalá, Nº58 es la más barata y su precio es de 12000 USD`

```python
casabarata = ds.loc[ds["price"].idxmin()]
print(f'La casa mas barata es {casabarata["address"]}, su precio es de {casabarata["price"]} USD')
```

#### Ejercicio 03. ¿Cuál es la casa más grande y la más pequeña del dataset? (★☆☆)

Imprime la dirección y el área de las casas seleccionadas. Por ejemplo:

`La casa más grande está ubicada en Calle Gran Vía, Nº38 y su superficie es de 5000 metros`

`La casa más pequeña está ubicada en Calle Mayor, Nº12 y su superficie es de 200 metros`

Este ejercicio es similar al anterior, pero buscamos las casas más grandes y pequeñas en base a su superficie.

```python
casagrande = ds.loc[ds["surface"].idxmax()]

casapequena = ds.loc[ds["surface"].idxmin()]

print(f'La casa mas grande está localizada en {casagrande["address"]} y su superficie es de {casagrande["surface"]} metros')
print(f'La casa más pequeña está localizada en {casapequena["address"]} y su superficie es de {casapequena["surface"]} metros')
```

#### Ejercicio 04. ¿Cuantas poblaciones contiene el dataset? (★☆☆)

Cuenta el número de poblaciones únicas en la columna 'level5' e imprime el nombre de las poblaciones separadas por coma. Por ejemplo:

`> print(populations)`

`population1, population2, population3, ...`

```python
poblaciones = ds["level5"].dropna().unique()

print("Número de poblaciones únicas:", ds["level5"].nunique())
print(", ".join(map(str, poblaciones)))
```

#### Ejercicio 07. ¿Cuál es la media de precios en la población de "Arroyomolinos (Madrid)"? (★★☆)

Imprima el valor obtenido de la columna level5.

```python
arroyomolinos = ds[ds["level5"] == "Arroyomolinos (Madrid)"]
media_precios = arroyomolinos["price"].mean()

print("La media de Arroyomolinos es:", media_precios)
```

#### Ejercicio 08. Trazar el histograma de los precios para la población de "Arroyomolinos (Madrid)" y explica qué observas (★★☆)

Imprime el histograma de los precios y escribe en la celda del Markdown un breve análisis del trazado.


```python
import matplotlib.pyplot as plt

arroyomolinos = ds[ds["level5"] == "Arroyomolinos (Madrid)"]

plt.figure(figsize=(10, 6))
plt.hist(arroyomolinos["price"], bins=20, edgecolor="black")
plt.title('Histograma de precios en Arroyomolinos (Madrid)')
plt.xlabel('Precio')
plt.ylabel('Frecuencia')
```

```python
# El histograma muestra cómo se distribuyen los precios de las viviendas en Arroyomolinos (Madrid).  
# Las barras aisladas hacia la derecha sugiere que hay casas mucho más caras que el resto 
```
